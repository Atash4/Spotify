```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
spotify_df = pd.read_csv('/kaggle/input/spotify-dataset/spotify_songs.csv')
```

```python
spotify_df.rename( columns = {'track_popularity':'popularity',
                                 'track_album_name':'track_name',
                                 'track_album_release_date':'release_date','track_artist':'artist'}, inplace = True)
```

```python
spotify_df.head(30000)
```

```python
spotify_df.info()
```

```python
spotify_df['duration'] = (spotify_df['duration_ms'] / 1000).round()
spotify_df.drop('duration_ms', inplace=True, axis=1)
spotify_df.duration.head()
```

```python
print(spotify_df.columns)
```

```python

spotify_df['duration'] = spotify_df['duration'] / 60

spotify_df['duration'] = spotify_df['duration'].round(2)

print(spotify_df[['duration']])
```

```python
ang = spotify_df.sample(int(0.006 * len(spotify_df)))
len(ang)
```

# Let's analyse the correlation between the popularity and acoustability utilizng visual tools such as matplotlib

```python
plt.figure(figsize=(10,6))
sns.regplot(data=ang, y='popularity', x='acousticness', color='g').set(title='Popularity vs Acousticness Correlation')
```

# Let's add another regression plot and analyzed the correlation between loudness and Energy

```python
plt.figure(figsize=(12,6))
sns.regplot(data=ang, y='loudness', x='energy', color='b').set(title='Loudness vs Energy Correlation')
```

# Retrieving the amount of songs released by each year

```python
spotify_df['release_date'] = pd.to_datetime(spotify_df['release_date'], errors='coerce')

spotify_df['years'] = spotify_df['release_date'].dt.year

sns.displot(data=spotify_df, x='years', discrete=True, aspect=3, height=6, kind='hist', color= 'cyan')

plt.title('Number of songs each year')

# Show the plot
plt.show()
```

# Now lets create a heat map of the pearson correlations of different framework of each song

```python

numeric_columns = spotify_df.select_dtypes(include=['number'])

# Calculating correlation matrix
cm = numeric_columns.corr(method='pearson')

# Plotting a heatmap
plt.figure(figsize=(14, 6))
sns.heatmap(cm, annot=True, fmt='.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linewidths=1, linecolor='Black')

plt.show()
```

# Plotting a bargraph to show which year has the most average minutes in songs

```python
spotify_df['years'] = spotify_df['years'].fillna(0)
spotify_df['years'] = spotify_df['years'].astype(int)
fig_dims = (18, 7)
fig, ax = plt.subplots(figsize=fig_dims)
spotify_df = spotify_df[spotify_df['years'] != 0]

sns.barplot(x='years', y='duration', data=spotify_df, ax=ax, errwidth=False)
plt.title('Years vs Duration')
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

# Now lets plot the average duration of songs in each genre

```python
average_duration_by_genre = spotify_df.groupby('playlist_genre')['duration'].mean().reset_index()
sns.barplot(x='duration', y='playlist_genre', data=average_duration_by_genre, palette='ocean')

plt.title('Average duration of songs in different Genres')
plt.xlabel('Average Duration (minutes)')
plt.ylabel('Genres')
plt.show()
```

# Retreiving data on which genres are the 'top 5 popular genres' 

```python
sns.set_style(style='darkgrid')
plt.figure(figsize=(12, 6))

# Replace 'spotify_df' with the actual name of your DataFrame
popular = spotify_df.sort_values('popularity', ascending=False).head(10)

# Create the bar plot
sns.barplot(y='playlist_genre', x='popularity', data=popular)

plt.title('Top 5 Genres by Popularity')
plt.xlabel('Popularity')
plt.ylabel('Genres')
plt.show()
```
